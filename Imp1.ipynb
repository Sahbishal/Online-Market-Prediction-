```python
# 4. Bar Chart: Average Positive Sentiment per Top 10 Companies (IMPROVED: Sorted)
# Group by company and get the average of positive news, then sort and take the top 10
top_positive = df.groupby('Symbol')['News - Positive Sentiment'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(8,5))
sns.barplot(x=top_positive.index, y=top_positive.values, palette='viridis')
plt.title("Top 10 Companies by Average Positive News")
plt.ylabel("Average Daily Positive News Count")
plt.xlabel("Company Symbol")
plt.tight_layout()
plt.savefig("bar_positive_sentiment.png")
plt.show()
```

```python
# 5. Pie Chart: Proportion of Positive vs Negative News
total_positive = df['News - Positive Sentiment'].sum()
total_negative = df['News - Negative Sentiment'].sum()
total_news = total_positive + total_negative

# Only create the chart if there is news data to avoid errors
if total_news > 0:
    plt.figure(figsize=(6,6))
    plt.pie([total_positive, total_negative], 
            labels=['Positive','Negative'], 
            autopct='%1.1f%%', 
            colors=['#4CAF50','#F44336'], 
            startangle=90)
    plt.title("Proportion of Positive vs Negative News")
    plt.tight_layout()
    plt.savefig("pie_sentiment.png")
    plt.show()
```

```python
# 6. Boxplot: Outliers in Close Prices (IMPROVED: Zoomed-in)
# -----------------------------
plt.figure(figsize=(6,4))
sns.boxplot(y=df['Close'], color='orange')
plt.title("Boxplot: Outliers in Close Prices")
plt.ylabel("Close Price (USD)")
plt.ylim(0, 1000) # Set y-axis limit to focus on the main data
plt.tight_layout()
plt.savefig("boxplot_close_zoomed.png")
plt.show()
```

```python
# 7. Line Chart: Trend of Close Price for One Company (e.g., AAPL)
# Make sure to handle potential missing values
company = 'AAPL'
company_df = df[df['Symbol']==company].sort_values('Date').dropna(subset=['Close'])

plt.figure(figsize=(10,4))
plt.plot(company_df['Date'], company_df['Close'], marker='', linestyle='-', color='#1f77b4', linewidth=1)
plt.title(f"Line Chart: Close Price Trend for {company}")
plt.xlabel("Date")
plt.ylabel("Close Price (USD)")
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.6) # Add a grid for better readability
plt.tight_layout()
plt.savefig("line_trend_aapl.png")
plt.show()
```

```python
# 8. Scatter Plot: Positive Sentiment vs Price Change (IMPROVED: With Regression Line)
# -----------------------------
plt.figure(figsize=(6,4))
# Use regplot to automatically add a linear regression line
sns.regplot(x='News - Positive Sentiment', y='Price_Change', data=df, 
            scatter_kws={'alpha':0.3, 's':10}, line_kws={'color':'red'})
plt.title("Scatter Plot: Positive News vs. Daily Price Change")
plt.xlabel("Positive News Count")
plt.ylabel("Price Change (USD)")
plt.tight_layout()
plt.savefig("scatter_sentiment_price_reg.png")
plt.show()
```
